Passo 2: Gerar Dados Fictícios
Vamos gerar um dataset fictício com as colunas timestamp, heart_rate, calories_burned e activity_type.

Código para Gerar Dados Fictícios

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Função para gerar dados fictícios
def gerar_dados_ficticios(num_registros):
    # Gerar timestamps
    timestamps = [datetime.now() - timedelta(minutes=i) for i in range(num_registros)]
    
    # Gerar dados de frequência cardíaca (bpm)
    heart_rate = np.random.randint(60, 180, size=num_registros)
    
    # Gerar dados de calorias queimadas
    calories_burned = np.random.randint(50, 500, size=num_registros)
    
    # Gerar tipos de atividade
    activity_types = np.random.choice(['Corrida', 'Ciclismo', 'Caminhada', 'Natação'], size=num_registros)
    
    # Criar DataFrame
    df = pd.DataFrame({
        'timestamp': timestamps,
        'heart_rate': heart_rate,
        'calories_burned': calories_burned,
        'activity_type': activity_types
    })
    
    return df

# Gerar dados fictícios com 100 registros
df_ficticio = gerar_dados_ficticios(100)

# Salvar os dados em um arquivo CSV
df_ficticio.to_csv('dados_desempenho_ficticio.csv', index=False)

# Exibir as primeiras linhas do DataFrame
print(df_ficticio.head())
```

Passo 3: Processar Dados
Depois de gerar os dados, vamos carregá-los e processá-los.

```python
# Carregar dados do CSV
df = pd.read_csv('dados_desempenho_ficticio.csv')

# Exibir as primeiras linhas do dataframe
print(df.head())
```

Passo 4: Criar o Dashboard
Vamos usar a biblioteca matplotlib para criar gráficos e visualizar os dados.

```python
import matplotlib.pyplot as plt

# Gráfico de frequência cardíaca ao longo do tempo
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['heart_rate'], label='Frequência Cardíaca')
plt.xlabel('Timestamp')
plt.ylabel('Frequência Cardíaca (bpm)')
plt.title('Frequência Cardíaca ao Longo do Tempo')
plt.legend()
plt.show()

# Gráfico de calorias queimadas ao longo do tempo
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['calories_burned'], label='Calorias Queimadas', color='orange')
plt.xlabel('Timestamp')
plt.ylabel('Calorias Queimadas')
plt.title('Calorias Queimadas ao Longo do Tempo')
plt.legend()
plt.show()
```
