```python
import pandas as pd

# Caminhos dos arquivos
file_paths = [
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/processed/indicadores.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/processed/indicadores2.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/processed/indicadores3.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/processed/indicadores4.csv"
]

# Carregar os arquivos em dataframes
dfs = [pd.read_csv(file) for file in file_paths]

# Exibir as primeiras linhas de cada dataframe para análise inicial
df_heads = [df.head() for df in dfs]
df_heads
```

```python
# Verificar as primeiras linhas dos arquivos brutos
raw_contents = {file: open(file, "r", encoding="utf-8").readlines()[:10] for file in file_paths}
raw_contents
```

```python
import pandas as pd

# Tentar abrir os arquivos com diferentes configurações de encoding e delimitador
encodings = ['utf-8', 'latin1', 'ISO-8859-1']
delimiters = [',', ';', '\t']

# Testar combinações
dataframes = {}
for file in file_paths:
    for encoding in encodings:
        for delimiter in delimiters:
            try:
                df = pd.read_csv(file, encoding=encoding, delimiter=delimiter, nrows=5)
                dataframes[(file, encoding, delimiter)] = df
            except Exception as e:
                dataframes[(file, encoding, delimiter)] = str(e)

dataframes
```
