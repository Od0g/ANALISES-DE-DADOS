```python
import pandas as pd

# Caminhos dos arquivos
files = [
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/raw/indicadores.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/raw/indicadores2.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/raw/indicadores3.csv",
    "/workspaces/ANALISES-DE-DADOS/AnalisedePCM/data/raw/indicadores4.csv"
]

# Carregar os arquivos e exibir informações básicas
dataframes = [pd.read_csv(file, encoding="utf-8", low_memory=False) for file in files]

# Verificar estrutura dos dados
info_list = [df.info() for df in dataframes]
head_list = [df.head() for df in dataframes]

info_list, head_list
```

```python
import chardet

# Detectar encoding dos arquivos
encodings = []
for file in files:
    with open(file, "rb") as f:
        result = chardet.detect(f.read(10000))  # Lendo uma amostra do arquivo
        encodings.append(result["encoding"])

encodings
```

```python
# Testar diferentes delimitadores
delimiters = [",", ";", "\t", "|"]

# Verificar os primeiros caracteres de cada arquivo para identificar o delimitador correto
samples = {}
for file in files:
    with open(file, "r", encoding="utf-8-sig") as f:
        line = f.readline()
        samples[file] = {delim: line.count(delim) for delim in delimiters}

samples
```

```python
# Carregar os arquivos corretamente usando ';' como delimitador
dataframes = [pd.read_csv(file, encoding="utf-8-sig", delimiter=";", low_memory=False) for file in files]

# Exibir informações básicas e as primeiras linhas
info_list = [df.info() for df in dataframes]
head_list = [df.head() for df in dataframes]

info_list, head_list
```

```python
# Carregar os arquivos corretamente usando ';' como delimitador
dataframes = [pd.read_csv(file, encoding="utf-8-sig", delimiter=";", low_memory=False) for file in files]

# Exibir informações básicas e as primeiras linhas
info_list = [df.info() for df in dataframes]
head_list = [df.head() for df in dataframes]

info_list, head_list
```

```python
# Concatenar os DataFrames para formar um único conjunto de dados
df_combined = pd.concat(dataframes, ignore_index=True)

# Verificar a estrutura do dataset combinado
df_combined.info(), df_combined.head()
```

```python
# Remover colunas completamente vazias
df_cleaned = df_combined.dropna(axis=1, how="all")

# Converter colunas de datas para formato datetime
date_columns = ["DATA_ABERTURA", "DATA_PARADA", "DATA_REINICIO", "DATA_FECHAMENTO", "DATA_ALTERACAO", "data.prog.inicio", "data.prog.fim", "data.real.inicio"]
for col in date_columns:
    if col in df_cleaned.columns:
        df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors="coerce", dayfirst=True)

# Exibir informações após a limpeza
df_cleaned.info(), df_cleaned.head()
```

```python
# Contar a quantidade de OS por setor
os_por_setor = df_cleaned["SETOR"].value_counts()

# Calcular o tempo médio de atendimento (diferença entre DATA_ABERTURA e DATA_FECHAMENTO)
df_cleaned["tempo_atendimento"] = (df_cleaned["DATA_FECHAMENTO"] - df_cleaned["DATA_ABERTURA"]).dt.total_seconds() / 3600  # em horas
tempo_medio_atendimento = df_cleaned["tempo_atendimento"].mean()

# Identificar problemas mais recorrentes
problemas_mais_comuns = df_cleaned["DESCRICAO"].value_counts().head(10)

os_por_setor, tempo_medio_atendimento, problemas_mais_comuns
```

Insights iniciais:
Setores com mais sistemas operacionais abertos:

Extrusão lidera com 10.895 ordens de serviço.

Impressão (3.332 SOs) e Rebobinadeira (2.056 OSs) também geram muitas conversões.

Tempo médio de atendimento:

Aproximadamente 1185 horas (quase 50 dias), indicando que alguns sistemas operacionais podem ficar abertos por longos períodos.

Tipos de problemas mais recorrentes:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo
sns.set_theme(style="whitegrid")

# Gráfico de OSs por setor (Top 10)
plt.figure(figsize=(12, 6))
os_por_setor.head(10).plot(kind="bar", color="royalblue")
plt.title("Top 10 Setores com Mais Ordens de Serviço")
plt.xlabel("Setor")
plt.ylabel("Quantidade de OSs")
plt.xticks(rotation=45)
plt.show()

# Gráfico de Tipos de Manutenção
plt.figure(figsize=(10, 5))
problemas_mais_comuns.plot(kind="bar", color="tomato")
plt.title("Tipos de Manutenção Mais Comuns")
plt.xlabel("Tipo de Manutenção")
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.show()
```

```python
# Tempo médio de atendimento por setor
tempo_medio_por_setor = df_cleaned.groupby("SETOR")["tempo_atendimento"].mean().sort_values(ascending=False)

# Contagem de preventivas x corretivas ao longo do tempo
df_cleaned["mes_ano"] = df_cleaned["DATA_ABERTURA"].dt.to_period("M")
manutencao_por_mes = df_cleaned.groupby(["mes_ano", "DESCRICAO"]).size().unstack().fillna(0)

# Tempo médio de atendimento por tipo de máquina
if "EQUIPAMENTO" in df_cleaned.columns:
    tempo_medio_por_equipamento = df_cleaned.groupby("EQUIPAMENTO")["tempo_atendimento"].mean().sort_values(ascending=False)
else:
    tempo_medio_por_equipamento = None

# Gráfico 1: Tempo médio de atendimento por setor
plt.figure(figsize=(12, 6))
tempo_medio_por_setor.plot(kind="bar", color="darkorange")
plt.title("Tempo Médio de Atendimento por Setor")
plt.xlabel("Setor")
plt.ylabel("Tempo Médio (horas)")
plt.xticks(rotation=45)
plt.show()

# Gráfico 2: Evolução de preventivas vs corretivas
plt.figure(figsize=(12, 6))
manutencao_por_mes[["Corretiva", "Preventiva"]].plot(kind="line", marker="o", figsize=(12, 6))
plt.title("Evolução das Manutenções Preventivas e Corretivas")
plt.xlabel("Mês/Ano")
plt.ylabel("Quantidade de OSs")
plt.xticks(rotation=45)
plt.legend(["Corretiva", "Preventiva"])
plt.show()

# Gráfico 3: Tempo médio de atendimento por equipamento (se disponível)
if tempo_medio_por_equipamento is not None:
    plt.figure(figsize=(12, 6))
    tempo_medio_por_equipamento.head(10).plot(kind="bar", color="purple")
    plt.title("Tempo Médio de Atendimento por Equipamento (Top 10)")
    plt.xlabel("Equipamento")
    plt.ylabel("Tempo Médio (horas)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Ajustar a categorização das OSs para identificar corretamente preventivas e corretivas
df_cleaned["tipo_manutencao"] = df_cleaned["DESCRICAO"].apply(
    lambda x: "Preventiva" if "preventiva" in str(x).lower() else "Corretiva" if "corretiva" in str(x).lower() else "Outros"
)

# Recalcular a contagem por mês com a nova categorização
manutencao_por_mes_corrigido = df_cleaned.groupby(["mes_ano", "tipo_manutencao"]).size().unstack().fillna(0)

# Gerar gráfico corrigido de preventivas vs corretivas
plt.figure(figsize=(12, 6))
manutencao_por_mes_corrigido[["Corretiva", "Preventiva"]].plot(kind="line", marker="o", figsize=(12, 6))
plt.title("Evolução das Manutenções Preventivas e Corretivas")
plt.xlabel("Mês/Ano")
plt.ylabel("Quantidade de OSs")
plt.xticks(rotation=45)
plt.legend(["Corretiva", "Preventiva"])
plt.show()
```

```python
# Substituir valores nulos por 'Outros' na coluna tipo_manutencao
df_cleaned["tipo_manutencao"] = df_cleaned["tipo_manutencao"].fillna("Outros")

# Remover linhas onde tempo_atendimento está nulo, pois é uma métrica essencial
df_cleaned = df_cleaned.dropna(subset=["tempo_atendimento"])

# Criar uma nova coluna para classificar manutenções como preventiva ou corretiva
df_cleaned["tipo_manutencao"] = df_cleaned["tipo_manutencao"].apply(
    lambda x: "Preventiva" if "PREVENTIVA" in str(x).upper() else "Corretiva"
)

# Verificar a distribuição dos tipos de manutenção após a correção
df_cleaned["tipo_manutencao"].value_counts()
```
